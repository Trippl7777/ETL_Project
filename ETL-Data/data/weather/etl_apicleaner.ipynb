```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime as dt
```

```python
atl_df = pd.read_csv("raw data/atl_api.csv")
atl_df = atl_df.rename(columns = {"Unnamed: 0":"index","Time":"Date (UNIX)"})
atl_df = atl_df.dropna()
atl_df["city"] = "atlanta"
atl_df["DateTime"] = pd.to_datetime(atl_df['Date (UNIX)'], unit = 's')
atl_df["Date"] = atl_df["DateTime"].dt.date
atl_df["Time"] = atl_df["DateTime"].dt.hour

#city, weather, date, time
atl_df = atl_df[["city","Weather","Date","Time"]]
atl_df = atl_df.rename(columns = {"Weather":"weather","Date":"date","Time":"time"})
atl_df = atl_df.drop_duplicates(subset = ["city","date","time"], keep = 'first')

atl_df.to_csv("atl_df.csv",index = False)
atl_df.head(-30)
```
